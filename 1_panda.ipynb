```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('resume_data.csv', encoding='utf-8-sig')
```

```python
data.head()
```

Drop Columns which is not required

```python
#Drop columns

updated_data = data.drop(columns=[
    'career_objective', 'educational_results', 'result_types',
    'professional_company_names', 'related_skils_in_job', 'locations',
    'extra_curricular_activity_types', 'extra_curricular_organization_names',
    'extra_curricular_organization_links', 'role_positions',
    'languages', 'proficiency_levels', 'certification_providers',
    'certification_skills', 'issue_dates', 'expiry_dates',
    'age_requirement', 'responsibilities.1', 'online_links','\ufeffjob_position_name',
    'address','company_urls'
], errors='ignore')
```

Unique values

```python
# See unique value counts per column
for col in updated_data.columns:
    print(f"\n{col}: {updated_data[col].nunique()} unique values")
    print(updated_data[col].value_counts().head(5))
```

```python
updated_data.info()
```

```python
print("Total rows:", len(updated_data))
print("Fully duplicate rows:", updated_data.duplicated().sum())

# Check duplicates based on key columns only
key_cols = ['skills', 'skills_required', 'matched_score']
print("Duplicates on key cols:", updated_data.duplicated(subset=key_cols).sum())
```
